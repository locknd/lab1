#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s):
    vowels = "aeiou"
    count = 0

    for ch in s.lower():
        if ch in vowels:
            count += 1

    return count
```

```python
# Тест 1. Обычное слово
assert count_vowels("hello") == 2
print("Тест 1 пройден")

# Тест 2. Все буквы — гласные
assert count_vowels("aeiou") == 5
print("Тест 2 пройден")

# Тест 3. Нет гласных
assert count_vowels("rhythm") == 0
print("Тест 3 пройден")

# Тест 4. Пустая строка
assert count_vowels("") == 0
print("Тест 4 пройден")

# Тест 5. Заглавные буквы
assert count_vowels("AEIOU") == 5
print("Тест 5 пройден")

# Тест 6. Смешанный регистр
assert count_vowels("ApplE") == 2
print("Тест 6 пройден")

print("Все тесты пройдены успешно")
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique(s):
    return len(s) == len(set(s))
```

```python
# Тест 1. Все символы уникальны
assert all_unique("zxc") is True
print("Тест 1 пройден")

# Тест 2. Есть повторяющиеся символы
assert all_unique("lenin") is False
print("Тест 2 пройден")

# Тест 3. Пустая строка
assert all_unique("") is True
print("Тест 3 пройден")

# Тест 4. Один символ
assert all_unique("1") is True
print("Тест 4 пройден")

# Тест 5. Повтор пробела
assert all_unique("a b c a") is False
print("Тест 5 пройден")

# Тест 6. С учетом регистра
assert all_unique("Aa") is True
print("Тест 6 пройден")

print("Все тесты пройдены успешно")
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n < 0:
        raise ValueError("n должно быть неотрицательным числом")
        
    iters = 0

    while n >= 10:
        i = 1
        for digit in str(n):
            i *= int(digit)
        n = i
        iters += 1

    return iters
```

```python
# Тест 1. Пример из условия
try:
    magic(-39)
    assert False, "Ожидалась ошибка для отрицательного числа"
except ValueError:
    print("Тест 1 на отрицательное число пройден")

# Тест 2. Однозначное число
assert magic(4) == 0
print("Тест 2 пройден")

# Тест 3. Пример из условия
assert magic(999) == 4
print("Тест 3 пройден")

# Тест 4. 25 -> 10 -> 0
assert magic(25) == 2
print("Тест 4 пройден")

# Тест 5. 77 -> 49 -> 36 -> 18 -> 8
assert magic(77) == 4
print("Тест 5 пройден")

# Тест 6. 10 -> 0
assert magic(10) == 1
print("Тест 6 пройден")

print("Все тесты пройдены успешно")
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")

    total = 0
    for p, t in zip(pred, true):
        total += np.square(p - t)

    return total / len(pred)
```

```python
# Тест 1. Полное совпадение
assert mse([1, 2, 3], [1, 2, 3]) == 0
print("Тест 1 пройден")

# Тест 2. Простой случай
# (1-1)^2 + (2-3)^2 + (3-5)^2 = 0 + 1 + 4 = 5
# 5 / 3
assert mse([1, 2, 3], [1, 3, 5]) == 5 / 3
print("Тест 2 пройден")

# Тест 3. Отрицательные числа
# (-1-1)^2 + (0-0)^2 = 4 + 0 = 4
# 4 / 2 = 2
assert mse([-1, 0], [1, 0]) == 2
print("Тест 3 пройден")

# Тест 4. Один элемент
assert mse([4], [1]) == 9
print("Тест 4 пройден")

# Тест 5. Ошибка при разной длине
try:
    mse([1, 2], [1])
    assert False, "Ожидалась ошибка для векторов разной длины"
except ValueError:
    print("Тест 5 пройден")

print("Все тесты пройдены успешно")
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors_str(n):
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    if n == 1:
        return "(1)"

    parts = []
    d = 2

    while d * d <= n:
        power = 0
        while n % d == 0:
            n //= d
            power += 1

        if power > 0:
            if power == 1:
                parts.append(f"({d})")
            else:
                parts.append(f"({d}**{power})")

        d += 1

    if n > 1:
        parts.append(f"({n})")

    return "".join(parts)
```

```python
# Тест 1. Пример из условия
assert prime_factors_str(86240) == "(2**5)(5)(7**2)(11)"
print("Тест 1 пройден")

# Тест 2. Простое число
assert prime_factors_str(13) == "(13)"
print("Тест 2 пройден")

# Тест 3. Степень одного простого числа
assert prime_factors_str(32) == "(2**5)"
print("Тест 3 пройден")

# Тест 4. Два множителя
assert prime_factors_str(12) == "(2**2)(3)"
print("Тест 4 пройден")

# Тест 5. Единица
assert prime_factors_str(1) == "(1)"
print("Тест 5 пройден")

# Тест 6. Ошибка для неположительного числа
try:
    prime_factors_str(0)
    assert False, "Ожидалась ошибка"
except ValueError:
    print("Тест 6 пройден")

print("Все тесты пройдены успешно")
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(num):
    if num <= 0:
        return "It is impossible"

    current = 0
    k = 0

    while current < num:
        k += 1
        current += k ** 2

    if current == num:
        return k
    return "It is impossible"
```

```python
# Тест 1. 1 = 1^2
assert pyramid(1) == 1
print("Тест 1 пройден")

# Тест 2. 5 = 1^2 + 2^2
assert pyramid(5) == 2
print("Тест 2 пройден")

# Тест 3. 14 = 1^2 + 2^2 + 3^2
assert pyramid(14) == 3
print("Тест 3 пройден")

# Тест 4. 30 = 1^2 + 2^2 + 3^2 + 4^2
assert pyramid(30) == 4
print("Тест 4 пройден")

# Тест 5. 2 не представляется как сумма 1^2 + ... + k^2
assert pyramid(2) == "It is impossible"
print("Тест 5 пройден")

# Тест 6. 55 = 1^2 + 2^2 + 3^2 + 4^2 + 5^2
assert pyramid(55) == 5
print("Тест 6 пройден")

# Тест 7. Неположительное число
assert pyramid(0) == "It is impossible"
print("Тест 7 пройден")

print("Все тесты пройдены успешно")
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    if n <= 0:
        raise ValueError("Число должно быть положительным")

    ch = [int(d) for d in str(n)]
    length = len(ch)

    if length % 2 == 1:
        mid = length // 2
        left_sum = sum(ch[:mid])
        right_sum = sum(ch[mid + 1:])
    else:
        mid1 = length // 2 - 1
        mid2 = length // 2
        left_sum = sum(ch[:mid1])
        right_sum = sum(ch[mid2 + 1:])

    return left_sum == right_sum
```

```python
# Тест 1. Нечетное число цифр, сбалансировано
# 12321 -> слева 1+2=3, справа 2+1=3
assert is_balanced(12321) is True
print("Тест 1 пройден")

# Тест 2. Нечетное число цифр, не сбалансировано
# 12341 -> слева 1+2=3, справа 4+1=5
assert is_balanced(12341) is False
print("Тест 2 пройден")

# Тест 3. Четное число цифр, сбалансировано
# 123321 -> средние цифры 3 и 3, слева 1+2=3, справа 2+1=3
assert is_balanced(123321) is True
print("Тест 3 пройден")

# Тест 4. Четное число цифр, не сбалансировано
# 123421 -> средние цифры 3 и 4, слева 1+2=3, справа 2+1=3
# стоп, это как раз сбалансировано, поэтому берём другой пример:
assert is_balanced(123451) is False  # слева 1+2=3, справа 5+1=6
print("Тест 4 пройден")

# Тест 5. Однозначное число
# слева и справа пусто, суммы равны 0
assert is_balanced(7) is True
print("Тест 5 пройден")

# Тест 6. Двузначное число
# две средние цифры — это и есть всё число, слева и справа пусто
assert is_balanced(42) is True
print("Тест 6 пройден")

# Тест 7. Ошибка для неположительного числа
try:
    is_balanced(0)
    assert False, "Ожидалась ошибка"
except ValueError:
    print("Тест 7 пройден")

print("Все тесты пройдены успешно")
```
